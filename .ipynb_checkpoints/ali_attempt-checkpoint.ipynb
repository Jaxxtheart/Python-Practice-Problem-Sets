```python
def check_palindrome(text):
    '''This function returns a boolean indicating whether the string is a palindrom or not'''
    half = int(len(text)/2) # half way length of text
    #lower casing text
    text = text.lower()
    
    # removing spaces
    text = text.replace(' ', '')
    
    # removing center word
    if len(text) % 2 == 1:
        text = text[:half] + text[half + 1:]
    
    # checking if first half is same as last half
    if text[:half] == text[:half-1:-1]: # second part is reversing string
        return True
    else:
        False
```
